### indexing과 활용

생활인구합계가 1000이상만 가져오고싶다면?

```python
sample["생활인구합계"] >= 1000
```

#### df[조건] | df.loc[조건] -> 조건에 맞는 데이터만 추출

```python
sample[sample["생활인구합계"] >= 1000]
```

#### 생활인구합계가 1000이상이고, 행정동코드가 11680610인 데이터만 추출 (and)

```python
condition = (sample["생활인구합계"] >= 1000) & (sample["행정동코드"] == 11680610)
condition
```

```python
sample[condition]
```

### 삭제

axis=0 행축을 의미 axis='row' -> 행축을 따라 위에서 아래로 탐색
axis=1 열축을 의미 axis='column' -> 열축을 따라 좌에서 우로 탐색

```python
sample.drop(["인구밀집지", "등록일자"], axis=1).head(2)
```

row삭제

```python
sample.drop(20170226, axis=0).head(2)
```
